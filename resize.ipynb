```python
import cv2
import numpy as np
import glob
import os
folders = glob.glob('/home/miglani/docface/lfwfeatures/*')
for folder in folders:
    #print(folder)
    for f in glob.glob(folder+'/*.jpg'):
        img = cv2.imread(f, cv2.IMREAD_UNCHANGED)
        #print('Original Dimensions : ',img.shape)
        a=img.shape
        name=os.path.basename(f)
        print(folder)
        if(a[0]<96 or a[1]<112):
            scale_percent = 220 # percent of original size
            width = int(img.shape[1] * scale_percent / 100)
            height = int(img.shape[0] * scale_percent / 100)
            dim = (width, height)
            # resize image
            resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)

            #print('Resized Dimensions : ',resized.shape)
            width = 112
            height = 96
            dim = (width, height)
            cropped = cv2.resize(resized, dim, interpolation = cv2.INTER_AREA)
            print('Resized Dimensions : ',cropped.shape)
            #cv2.imshow("Resized image", resized)
            cv2.imwrite(os.path.join(folder ,name),cropped)
        else:
            width = 112
            height = 96
            dim = (width, height)
            cropped = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
            print('Resized Dimensions : ',cropped.shape)
            #cv2.imshow("Resized image", resized)
            cv2.imwrite(os.path.join(folder ,name),cropped)
        
    cv2.waitKey(0)
    cv2.destroyAllWindows()
 
```
